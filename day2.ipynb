```python
# Import NumPy library
import numpy as np

# Create a 3D matrix A (2 x 2 x 2)
A = np.array([
    [[1, 2],
     [3, 4]],

    [[5, 6],
     [7, 8]]
])

B = np.array([10, 20])
result = A * B


print("Matrix A:")
print(A)

print("\nArray B:")
print(B)

print("\nResult after broadcasting multiplication:")
print(result)
```

```python
import pandas as pd

# Create the DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000],
    "Department": ["HR", "Finance", "Tech", "Tech", "HR"],
    "Experience": [2, 5, 8, 10, 12]
}

df = pd.DataFrame(data)

# Sort by Department (ascending) and Salary (descending)
sorted_df = df.sort_values(
    by=["Department", "Salary"],
    ascending=[True, False]
)

# Display the top 5 rows
print(sorted_df.head(5))
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

titanic_data = pd.read_csv(url)

# Average Fare by Passenger Class
print(titanic_data.groupby('Pclass')['Fare'].mean())

# Maximum Age by Embarked Port
print(titanic_data.groupby('Embarked')['Age'].max())

# Survival Rate by Gender
print(titanic_data.groupby('Sex')['Survived'].mean())

# Survival Rate by Passenger Class
print(titanic_data.groupby('Pclass')['Survived'].mean())
```

```python
import pandas as pd

# Load Housing Dataset
url = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv"

housing_data = pd.read_csv(url)

# Check missing values
print(housing_data.isnull().sum())

# Replace missing values in 'total_bedrooms' with the column mean
housing_data['total_bedrooms'] = housing_data['total_bedrooms'].fillna(
    housing_data['total_bedrooms'].mean()
)

# Verify missing values are replaced
print("\nAfter replacing missing values:")
print(housing_data.isnull().sum())
```
